# Change CSV to Fixed Width Txt

## Define Widths

```python
# change list to dict
w = { i : widths[i] for i in range(0, len(widths) ) }
w
```

## Read CSV File to Convert

```python
importfilename = "TESTLouisianaOGBQualifiedFundedTimePeriods_2020-05-01.csv"
with open(importfilename, 'r') as read_csv:
    csv_reader = csv.reader(read_csv, delimiter=',')
    list_of_rows = list(csv_reader)
    list_of_strings = []
    for row in list_of_rows:
        list_of_strings.append(','.join(row))
```

```python
list_of_rows
```

```python
list_of_strings
```

## Export to Fixed Width Txt File

```python
# write file with fixed with, left adjusting
exportfilename = "file" + ".txt"
with open(exportfilename,"w") as f:
    for line in list_of_strings:
        for col_nr, col in enumerate(line.strip().split(",")):
            # the :<{w[col_nr]}} - part is left-adjusting to certain width 
            f.write(f"{col:<{w[col_nr]}}") 
        f.write("\n")
```
